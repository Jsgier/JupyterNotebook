## Example 1.3.31:  
(a) Find the load reflection coefficient, the input impedance, and VSWR in the transmission line shown in 1.3.8a. The length is $\frac{\lambda}{8}$ and its characteristic impedance $Z_{0}$ is 50 $\Omega$  
(b) Evaluate V($\frac{\pi}{8}$), I($\frac{\pi}{8}$), P($\frac{\pi}{8}$), I(0), V(0), P(0)  
(c) Find the length in centimeters of the $\frac{\pi}{8}$ transmission line at f = 1GHz

![Figure 1.3.8](/notebooks/Jupyter_Images/Figure1.3.8.jpg)

### Variables:
$\Gamma_{0}$ - load reflection coefficient  
$Z_{0}$ - characteristic impedance  
$Z_{in}(\lambda/8)$ - input impedance  
$V_{SWR}$ - standing wave ratio  
$\beta$ - propogation constant  
$f$ - frequency  
$\lambda$ - wavelength  
$c$ - speed of light, 2.998E8 m/s  

### Relevant Equations:

 Equation &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp;&nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp;             |  Number 
 :-------------------------------------- | ------------------ 
 $\lambda = \frac{2\pi}{\beta}$     | (1.3.29)       
 $\lambda = v_{p}T$     | (1.3.31)       
 $\Gamma_{0} = \frac{Z_{L} - Z_{0}}{Z_{L} + Z_{0}}$  | (1.3.39)        
 $V_{SWR} = \frac{1 + \lvert \Gamma_{0} \rvert}{1 - \lvert \Gamma_{0} \rvert}$   | (1.3.44)   
 $Z_{IN}(d) = Z_{0} \frac{Z_{L} + jZ_{0}\tan(\beta d)}{Z_{0} + jZ_{L}\tan(\beta d)}$   | (1.3.44) 

```python
import numpy as np
import matplotlib.pyplot as plt
from cmath import polar
#load reflection coefficient
zL = np.complex(50, 50)
#characteristic impedance
z0 = np.complex(50, 0)
#calculate the load reflection coefficient gamma
G0 = (zL - z0)/(zL + z0)
#calculate VSWR
VSWR = (1+np.absolute(G0)) / (1 - np.absolute(G0))
#calculate input impedance
#note: Beta*d = pi/4, and tan(pi/4) = 1
zin = z0 * (np.complex(zL,z0))/(np.complex(z0, zL))
absZin = np.absolute(zin)

#print the results
print("VSWR = {:1.3} \nLoad Reflection Coefficient = {:1.3}".format(VSWR, G0))
print("Magnitude Reflection Coefficient = {:1.3} \n".format(np.absolute(G0)))
print("Zin = {:1.3}\nAbsolute Zin ={:1.3}".format(zin,absZin))

#rectangular plot of complex reflection coefficient
plt.arrow(0, 0, G0.real, G0.imag, length_includes_head = True, head_width = 0.02, shape = 'full')
plt.xlim(0, 0.5)
plt.ylim(0, 0.5)
plt.ylabel('Imaginary')
plt.xlabel('Real')
plt.title("Reflection Coefficient")
plt.grid(True, which='both')
plt.show()

#rectangular plot of complex load impedance
plt.arrow(0, 0, zL.real, zL.imag, length_includes_head = True, head_width = 5, shape = 'full')
plt.xlim(0, 100)
plt.ylim(0, 100)
plt.ylabel('Imaginary')
plt.xlabel('Real')
plt.title("Load Impedance")
plt.grid(True, which='both')
plt.show()

#rectangular plot of complex input impedance
plt.arrow(0, 0, zin.real, zin.imag, length_includes_head = True, head_width = 5, shape = 'full')
plt.xlim(0, 150)
plt.ylim(-150, 0)
plt.ylabel('Imaginary')
plt.xlabel('Real')
plt.title("Input Impedance at 1/8 Wavelength")
plt.grid(True, which='both')
plt.show()


```
